### What are 'if else' statements in Python?

'If else' statements are used for making decisions in your code. They allow your program to execute different blocks of code based on whether a certain condition is true or false.

*   **`if`**: The code inside the `if` block runs if the condition is `True`.
*   **`else`**: The code inside the `else` block runs if the `if` condition is `False`.

```python
# Example of an 'if else' statement

weather = "sunny"

if weather == "sunny":
    print("It's a beautiful day! Let's go for a walk.")
else:
    print("It might be cloudy or rainy. Grab an umbrella!")


# Another example with numbers
score = 75

if score >= 60:
    print("You passed the test!")
else:
    print("You need to study more.")
```

### Number Guessing Game

I'll pick a random number between 0 and 100. Try to guess it! I'll tell you if you're 'hot' (within 10 of the number) or 'cold'.

```python
import random

# Generate a random number between 0 and 100 (inclusive)
target_number = random.randint(0, 100)

# You can uncomment the line below for debugging to see the target number
# print(f"[DEBUG] The target number is: {target_number}")

print("I have picked a secret number between 0 and 100. Can you guess it?")
```

```python
guess = None
attempts = 0

while guess != target_number:
    try:
        guess_input = input("Enter your guess: ")
        guess = int(guess_input)
        attempts += 1

        if not (0 <= guess <= 100):
            print("Please guess a number between 0 and 100.")
            continue

        if guess == target_number:
            print(f"Congratulations! You guessed the number {target_number} in {attempts} attempts!")
        else:
            difference = abs(target_number - guess)
            if difference <= 10:
                print("Hot!")
            else:
                print("Cold!")
    except ValueError:
        print("Invalid input. Please enter a whole number.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
